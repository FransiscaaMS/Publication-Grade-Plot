---------------------------------

**Langkah 1 : Mendapatkan Top 5 major category**

Dengan cara mengambil kolom yang diperlukan (major_category), menghitung jumlahnya, dan menampilkan 5 teratas. Sehingga diperoleh hasil Top 5 major category seperti sebagai berikut.

```python
df['major_category'].value_counts().head(5)
```

**Langkah 2 : Mengelompokkan data pada kolom major_category dan year pada dataframe df**

Dengan cara menggunakan condition pada dataframe df yang memiliki year > 2007 dan groupby untuk melakukan pengelompokan data. Kemudian, jumlahkan masing-masing major_categori dengan sum().

```python
major = df[(df['year'] > 2007)].groupby(['major_category','year'])['Counted'].sum()
major
```

**Langkah 3 : Mengelompokkan data berdasarkan Top 5 major_category**

Dengan cara mendefinisikan variabel baru. Nanti akan diketahui urutan tahun dan banyaknya masing-masing kejahatan pada Top 5 major_category (Theft and Handling, Violence Against the Person, Criminal Damage, Drugs, Burglary) yang dilakukan pada tahun tersebut.

```python
Tehand = major['Theft and Handling']
Tehand
```

```python
Violence = major['Violence Against the Person']
Violence
```

```python
Crimdam = major['Criminal Damage']
Crimdam
```

```python
Drugs = major['Drugs']
Drugs
```

```python
Bgry = major['Burglary']
Bgry
```

**Langkah 4 : Mengonvert menjadi data frame baru**

Dengan cara menggunakan function **.to_frame()**, kemudian rename nama kolom 'Counted' menjadi masing-masing kejahatan pada Top 5 major_category (Theft and Handling, Violence Against the Person, Criminal Damage, Drugs, Burglary).

```python
dfTehand = Tehand.to_frame()
dfTehand.rename(columns={'Counted': 'Theft and Handling'}, inplace=True)
dfTehand.index.name=None
dfTehand
```

```python
dfViolence = Violence.to_frame()
dfViolence.rename(columns={'Counted': 'Violence Against the Person'}, inplace=True)
dfViolence.index.name=None
dfViolence
```

```python
dfCrimdam = Crimdam.to_frame()
dfCrimdam.rename(columns={'Counted': 'Criminal Damage'}, inplace=True)
dfCrimdam.index.name=None
dfCrimdam
```

```python
dfDrugs = Drugs.to_frame()
dfDrugs.rename(columns={'Counted': 'Drugs'}, inplace=True)
dfDrugs.index.name=None
dfDrugs
```

```python
dfBgry = Bgry.to_frame()
dfBgry.rename(columns={'Counted': 'Burglary'}, inplace=True)
dfBgry.index.name=None
dfBgry
```

**Langkah 5: Menggabungkan masing-masing dataframe menjadi satu**

Dengan cara membuat variabel baru dan menggunakan function **pd.concat([])**. Sehingga diperoleh urutan tahun, dan jumlah masing-masing kejahatan pada Top 5 major_category (Theft and Handling, Violence Against the Person, Criminal Damage, Drugs, Burglary).

```python
df_top5 = pd.concat([dfTehand,dfViolence,dfCrimdam,dfDrugs,dfBgry], axis=1)
df_top5
```

**Langkah 6 : Membuat Line Plot Top 5 major_category**

Dengan cara menggunakan function **.plot()** dan menambahkan ornamen lain seperti x label, y label, dan judul. Hasilnya berupa diagram garis sebagai berikut.

```python
df_top5.plot() # memanggil plot line
plt.rcParams['figure.figsize'] = [12,10] #mengubah ukuran canvas plot line
plt.xlabel('Year') #menyeting label x
plt.ylabel('Number of Crimes') #menyeting label y
plt.title('Top 5 types and number of crimes in london in 2007-2016') #judul plot line
plt.show() #menampilkan
```

**Langkah 7 : Membuat Area Plot Top 5 major_category**

Dengan menggunakan function **.plot(kind='area')** dan menambahkan  ornamen lain seperti x label, y label, dan judul. Hasilnya berupa diagram area sebagai berikut.

```python
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='area',
            stacked=False,
            figsize=(15,10),
            )

plt.title("Top 5 major crime in London 2008 - 2016")
plt.ylabel('Number of Criminal')
plt.xlabel('Years')

plt.show()
```

**Langkah 2: Mengelompokkan data pada dataframe df (Kota Wandsworth)**

```python
W = (df[(df['major_category'] == 'Criminal Damage') & (df['City'] == 'Wandsworth')].groupby('year')['Counted'].sum())
W
```

**Langkah 3: Mengelompokkan data pada dataframe df (Kota Greenwhich)**

Dengan cara menggunakan condition pada dataframe df yang memiliki major_category = Criminal Damage, City = Greenwich, dan groupby (year) untuk melakukan pengelompokan data. Kemudian, jumlahkan kasus Criminal Damage dengan sum().

```python
G = (df[(df['major_category'] == 'Criminal Damage') & (df['City'] == 'Greenwich')].groupby('year')['Counted'].sum())
G
```
